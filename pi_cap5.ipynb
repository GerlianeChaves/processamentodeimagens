#Escala

```python
#escala
img = cv2.imread('/gdrive/My Drive/Colab Notebooks/ProcessamentoImagem/Fig0236aletter_T-55eae8251be7404fab5c29fb96fbcb54.tif', 255)
height = int(img.shape[0] * (200/100))
width = int(img.shape[1]* (200/100))
dim = (width, height)

#interpolcao
#vizinho mais prox
img_resize_vizinho = cv2.resize(img, dim, interpolation = cv2.INTER_NEAREST) #vizinho mais proximo
img_resize_vizinho = cv2.putText(img_resize_vizinho, 'Figura 2: Interpolacao do vizinho mais proximo', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))

print('Vizinho  mais prox')
cv2_imshow(img_resize_vizinho)
print(img_resize_vizinho.shape)

#bilinear
img_resize_bil = cv2.resize(img, dim, interpolation = cv2.INTER_LINEAR) 
img_resize_bil = cv2.putText(img_resize_bil, 'Figura 3: Interpolacao bilinear', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255))

print('bilinear')
cv2_imshow(img_resize_bil)
print(img_resize_bil.shape)

#bicubica
img_resize_bic = cv2.resize(img, dim, interpolation = cv2.INTER_CUBIC) 
img_resize_bic = cv2.putText(img_resize_bic, 'Figura 3: Interpolacao bilinear', (350,350), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))
print('Bicubica')
cv2_imshow(img_resize_bic)
print(img_resize_bic.shape)

#plot
plt.figure()
plt.subplot(1, 3, 1) # Primeiro painel de uma grade 1x3
plt.imshow(img_resize_vizinho)
plt.title('Vizinho mais proximo')
plt.subplot(1, 3, 2) # Segundo painel de uma grade 1x3
plt.imshow(img_resize_bil)
plt.title('Bilinear')
plt.subplot(1, 3, 3) # Terceiro painel de uma grade 1x3
plt.imshow(img_resize_bic)
plt.title('bicubica')
plt.show() # Exibe a janela com as imagens
```

Aqui podemos observar que podemos alterar o valor da escala da imagem e que dentre as interpolações aplicadas, a bicubica apresenta um  efeito com maior nitidez de borda quando comparada a bilinear e que ambas são mais suaves na borda quando comparadas com a do vizinho mais próximo, que apresenta um efeito pixado.

#Rotação

```python
#rotacao
img = cv2.imread('/gdrive/My Drive/Colab Notebooks/ProcessamentoImagem/Fig0236aletter_T-55eae8251be7404fab5c29fb96fbcb54.tif', 255)
height = int(img.shape[0] * (100/100))
width = int(img.shape[1]* (100/100))
dim = (width, height)

ponto = (width / 2, height / 2) #ponto no centro da figura
rotacao = cv2.getRotationMatrix2D(ponto, 45, 1.0)
rotacionado = cv2.warpAffine(img, rotacao, (width, height))
cv2_imshow(rotacionado)

#interpolcao
#vizinho mais prox
img_resize_vizinho = cv2.resize(rotacionado, dim, interpolation = cv2.INTER_NEAREST) #vizinho mais proximo
img_resize_vizinho = cv2.putText(img_resize_vizinho, 'Figura 2: Interpolacao do vizinho mais proximo', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))

print('Vizinho  mais prox')
cv2_imshow(img_resize_vizinho)
print(img_resize_vizinho.shape)

#bilinear
img_resize_bil = cv2.resize(rotacionado, dim, interpolation = cv2.INTER_LINEAR) 
img_resize_bil = cv2.putText(img_resize_bil, 'Figura 3: Interpolacao bilinear', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255))

print('bilinear')
cv2_imshow(img_resize_bil)
print(img_resize_bil.shape)

#bicubica
img_resize_bic = cv2.resize(rotacionado, dim, interpolation = cv2.INTER_CUBIC) 
img_resize_bic = cv2.putText(img_resize_bic, 'Figura 3: Interpolacao bilinear', (350,350), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))
print('Bicubica')
cv2_imshow(img_resize_bic)
print(img_resize_bic.shape)

#plot
plt.figure()
plt.subplot(1, 3, 1) # Primeiro painel de uma grade 1x3
plt.imshow(img_resize_vizinho)
plt.title('Vizinho mais proximo')
plt.subplot(1, 3, 2) # Segundo painel de uma grade 1x3
plt.imshow(img_resize_bil)
plt.title('Bilinear')
plt.subplot(1, 3, 3) # Terceiro painel de uma grade 1x3
plt.imshow(img_resize_bic)
plt.title('bicubica')
plt.show() # Exibe a janela com as imagens
```

Aqui realizamos a rotação da imagem a partir do centro desta. Novamente, o método de interpolação do vizinho mais proximo é o que apresenta efeito mais pixado.

#Translação

```python
#translacao
img = cv2.imread('/gdrive/My Drive/Colab Notebooks/ProcessamentoImagem/Fig0236aletter_T-55eae8251be7404fab5c29fb96fbcb54.tif', 255)
height = int(img.shape[0] * (100/100))
width = int(img.shape[1]* (100/100))
dim = (width, height)

desloc = np.float32([[1, 0, 45], [0, 1, 50]])
deslocado = cv2.warpAffine(img, desloc, (width, height))
cv2_imshow(deslocado)

#interpolcao
#vizinho mais prox
img_resize_vizinho = cv2.resize(deslocado, dim, interpolation = cv2.INTER_NEAREST) #vizinho mais proximo
img_resize_vizinho = cv2.putText(img_resize_vizinho, 'Figura 2: Interpolacao do vizinho mais proximo', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))

print('Vizinho  mais prox')
cv2_imshow(img_resize_vizinho)
print(img_resize_vizinho.shape)

#bilinear
img_resize_bil = cv2.resize(deslocado, dim, interpolation = cv2.INTER_LINEAR) 
img_resize_bil = cv2.putText(img_resize_bil, 'Figura 3: Interpolacao bilinear', (30,900), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255))

print('bilinear')
cv2_imshow(img_resize_bil)
print(img_resize_bil.shape)

#bicubica
img_resize_bic = cv2.resize(deslocado, dim, interpolation = cv2.INTER_CUBIC) 
img_resize_bic = cv2.putText(img_resize_bic, 'Figura 3: Interpolacao bilinear', (350,350), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))
print('Bicubica')
cv2_imshow(img_resize_bic)
print(img_resize_bic.shape)

#plot
plt.figure()
plt.subplot(1, 3, 1) # Primeiro painel de uma grade 1x3
plt.imshow(img_resize_vizinho)
plt.title('Vizinho mais proximo')
plt.subplot(1, 3, 2) # Segundo painel de uma grade 1x3
plt.imshow(img_resize_bil)
plt.title('Bilinear')
plt.subplot(1, 3, 3) # Terceiro painel de uma grade 1x3
plt.imshow(img_resize_bic)
plt.title('bicubica')
plt.show() # Exibe a janela com as imagens
```
